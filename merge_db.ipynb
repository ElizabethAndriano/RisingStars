# Procesamiento de datos

Se eligen las columnas que se consideran necesarias para un primer modelo

```python
cols = ['Number of Authors','Scopus Author Ids','Year','Citations',
        'Field-Weighted Citation Impact',
        'Field-Weighted Outputs in Top Citation Percentiles, per percentile',
        'CiteScore (publication year)','SNIP (publication year)']
pub1 = pub[cols]
pub1.head()
```

```python
pub1.isna().sum().sort_values(ascending=False)
```

```python
pub1.dropna(inplace=True)
```

Como el enfoque es hacia los autores, con la base de datos de publicaciones se creará una nueva base de datos con los autores por separado

```python
px.box(pub1,x='Number of Authors')
```

Hay muchos outliers que pueden afectar el rendimiento del modelo, entonces se eliminarán antes de dividir la base de datos por autores

```python
Q1 = np.percentile(pub1['Number of Authors'],25)
Q3 = np.percentile(pub1['Number of Authors'],75)
IQR = Q3 - Q1
pub2 = pub1[pub1['Number of Authors'] <= Q3 + IQR * 1.5]

px.box(pub2,'Number of Authors')
```

```python
print('Datos perdidos:',pub1['Number of Authors'].size - pub2['Number of Authors'].size)
```

```python
autores = []
for auths in pub2['Scopus Author Ids']:
    autores.append(str(auths).split('| '))

pub2['Scopus Author Ids'] = autores
```

```python
pub2.head()
```

```python
pub2 = pub2.explode('Scopus Author Ids')
pub2.head()
```

```python
pub2['Publications'] = 1
pub2.head()
```

```python
metrics = {'Year':['max','min'],'Citations':[lambda x: list(x),'sum'],'Field-Weighted Citation Impact': 'mean',
           'Field-Weighted Outputs in Top Citation Percentiles, per percentile':'mean',
           'Publications':'sum','CiteScore (publication year)':'mean',
           'SNIP (publication year)':'mean'}
pub3 = pub2.groupby(by='Scopus Author Ids').agg(metrics)
pub3.head()
```

```python
pub3['Citations per Publication'] = pub3['Citations']['sum'] / pub3['Publications']['sum']
pub3.head()
```

```python
hIndex = []

for citations in pub3['Citations']['<lambda_0>']:
    citations.sort(reverse=True)
    calc = 0
    for j in citations:
        if j >= calc+1:
            calc += 1
        else: break
    hIndex.append(calc)

pub3['h-index'] = hIndex
pub3.head()
```

```python
# revisar que el h-index si aparezca bien calculado
pub3[pub3['Publications']['sum']> 10].head()
```

Ya que se cuenta con la misma información que la base de datos de autores, se almacena este nuevo dataframe en un archivo csv o excel para poder manejarlo mejor y combinarlo con la base de datos de autores

```python
pubAuth = pub3.copy()
pubAuth.reset_index(inplace=True)
pubAuth.head()
```

```python
pubAuth.drop([('Citations','<lambda_0>'),('Publications','sum')],axis=1,inplace=True)
pubAuth.columns = ['_'.join(col) for col in pubAuth.columns.values]
pubAuth.head()
```

```python
rename = {'Year_max':'Most recent publication',
          'Year_min':'Oldest publication (since 1996)',
          'Scopus Author Ids_':'Scopus author ID',
          'Citations_sum':'Citations',
          'Field-Weighted Citation Impact_mean':'Field-Weighted Citation Impact',
          'Field-Weighted Outputs in Top Citation Percentiles, per percentile_mean':
          'Output in Top 10% Citation Percentiles (field-weighted)',
          'CiteScore (publication year)_mean': 'CiteScore',
          'SNIP (publication year)_mean':'SNIP',
          'Citations per Publication_':'Citations per Publication',
          'h-index_':'h-index'}
pubAuth.rename(columns=rename,inplace=True)
pubAuth.head()
```

```python
pubAuth.to_excel('PublicationAuthorsMetrics.xlsx')
```
