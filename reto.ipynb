Se eliminan los datos faltantes de las columnas que tienen menos del 5% de datos faltantes

```python
Perc5 = pub.index.size * 0.05
drop = [cname for cname in pub.columns if 0 < pub[cname].isna().sum() < Perc5]

pub.dropna(subset=drop,inplace=True)
```

Se eligen las columnas que no se consideran necesarias para el modelo

```python
drop = ['Volume','Issue','Pages','Article number','ISSN','PubMed ID',
        'Sustainable Development Goals (2022)','Country/Region','Open Access',
        'Scopus Author ID Single Author','Scopus Author ID Corresponding Author',
        'Scopus Author ID Last Author','Scopus Author ID First Author']
pub.drop(columns=drop,inplace=True)
```
